```python
import pandas as pd
import glob

df = pd.DataFrame(columns=["path", "target"])
for fn in glob.glob("./clothes/tshirt/Tshirt_*.jpg"):
    s = pd.Series([fn, 0], index=["path", "target"])
    df = df.append(s, ignore_index=True)
for fn in glob.glob("./clothes/shirt/Shirt_*.jpg"):
    s = pd.Series([fn, 1], index=["path", "target"])
    df = df.append(s, ignore_index=True)
for fn in glob.glob("./clothes/suit/Suit_*.jpg"):
    s = pd.Series([fn, 2], index=["path", "target"])
    df = df.append(s, ignore_index=True)
for fn in glob.glob("./clothes/dress/Dress_*.jpg"):
    s = pd.Series([fn, 3], index=["path", "target"])
    df = df.append(s, ignore_index=True)
for fn in glob.glob("./clothes/vest/Vest_*.jpg"):
    s = pd.Series([fn, 4], index=["path", "target"])
    df = df.append(s, ignore_index=True)
df
```

```python
from keras.applications.vgg16 import preprocess_input
from keras.preprocessing.image import load_img
import numpy as np
imglist = []
for i in df["path"]:
    print(i)
    img = load_img(i, target_size=(224, 224))
    imglist.append(preprocess_input(np.array(img)))
```

```python
# import numpy as np
# x = np.array(imglist)

# x_train = x.reshape(x_train.shape[0], 224, 224, 3)
# print("x_train.shape={}".format(x_train.shape))
```

```python
from sklearn.model_selection import train_test_split
import numpy as np
x = np.array(imglist)
y = np.array(df["target"])
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from keras.utils.np_utils import to_categorical
y_train_clothes = to_categorical(y_train)
y_test_clothes = to_categorical(y_test)
```

```python
# trans = ["Tshirt", "Shirt", "Suit", "Dress", "Vest"]
```

```python
import keras
from keras.applications.resnet50 import ResNet50
rn = keras.applications.resnet50.ResNet50(include_top=False, 
                                     weights='imagenet',
                                     input_shape=(224, 224, 3))
```

```python
# x_train = x_train.reshape(x_train.shape[0], 224, 224, 3).astype('float32')
# x_test = x_test.reshape(x_test.shape[0], 224, 224, 3).astype('float32')

# print("x_train.shape={}".format(x_train.shape))
# print("x_test.shape={}".format(x_test.shape))
```

```python
from keras.models import Model
from keras.layers import BatchNormalization, Flatten, Dense, Dropout
for l in rn.layers:
    l.trainable = False
# ä¸èª¿æ•´å·ä¾†çš„åƒæ•¸,å³CNNä¸è¨“ç·´
x = BatchNormalization()(rn.output)
x = Flatten()(x)
#def test():
#    return print
#test()("hello"),å›žå‚³å‡½æ•¸å¾Œé¢å†æŽ¥ä¸€å€‹å‡½æ•¸
# x = Dense(1024, activation="relu")(x)
# x = Dropout(0.25)(x)
x = Dropout(0.5)(x)
# x = Dense(512, activation="relu")(x)
# x = Dense(256, activation="relu")(x)
# x = Dropout(0.5)(x)
# x = Dense(128, activation="relu")(x)
# x = Dropout(0.5)(x)
# ç¬¬ä¸€å€‹é¸æ“‡: 1 activation="sigmoid"
# ç¬¬äºŒå€‹é¸æ“‡: 2 activation="softmax"
x = Dense(5, activation="softmax")(x)
model = Model(inputs=rn.input, outputs=x)
model.summary()
```

```python
# optimizeræ¢¯åº¦ä¸‹é™æ–¹å¼,é¸æ“‡adamå„ªåŒ–å™¨ 
model.compile(loss="categorical_crossentropy", 
              optimizer="adam", 
              metrics=["accuracy"])
```

```python
model.fit(x_train, y_train_clothes,
        batch_size=200, epochs=10, 
        validation_split=0.1, verbose=2)
```

```python
model.evaluate(x_test, y_test_clothes)
```
